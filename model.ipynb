```python
# Importing important packages
import os
import sys

# Data Wrangling and manipulation
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn.model_selection import train_test_split

import nltk
from nltk.tokenize import word_tokenize

```

```python
df = pd.read_csv('cleaned_Resume.csv')
```

# Exploratory Data Analysis : check word frequency


```python
all_records = ''

# combining records into one string
for record in df['resume']:
    all_records += record
```

```python
all_tokens = word_tokenize(all_records) # tokenize word in list
word_freq = nltk.FreqDist(all_tokens)
mostcommon = word_freq.most_common(30)
common = pd.DataFrame(mostcommon)
common.rename(columns={0:'word', 1:'frequency'}, inplace = True)
```

```python
plt.xticks(rotation=90)
plt.title('Most common words')
sns.barplot(common, x='word', y='frequency')
```

# feature extraction : BoW, TF-IDF

```python

# Creating the Bag of Words model
word2count = {}
for resume in df['resume']:
    words = nltk.word_tokenize(resume)
    for word in words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] += 1
```
